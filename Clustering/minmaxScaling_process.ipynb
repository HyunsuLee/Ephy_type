# Data normalize for ANN(tensorflow)

* normalize 하지 않고, ANN model은 적용하면 step이 진행됨에 따라 cost가 다시 커지는 경향이 있음.
* 이는 learning rate나 weight initialization 최적화로 해결이 안됨.
* input node로 들어오는 feature별로 data의 scale이 달라 그럴 가능성이 큼.
* scale이 다르면 cost function의 optima를 향하는 길이 균등하지 않고, 기울어(skew)지기 때문.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cretrainX = np.loadtxt("cretrainX.csv", delimiter = ",")
cretestX = np.loadtxt("cretestX.csv", delimiter = ",")
```

```python
plt.scatter(cretrainX[:,0], cretrainX[:,1])
```

```python
from sklearn import preprocessing
```

```python
minmax_scale = preprocessing.MinMaxScaler().fit(cretrainX)
cretrainX_minmax = minmax_scale.transform(cretrainX)
```

```python
plt.scatter(cretrainX_minmax[:,0], cretrainX_minmax[:,1])
```

```python
cretestX_minmax=minmax_scale.transform(cretestX)
```

```python
plt.scatter(cretestX[:,0], cretestX[:,1])
```

```python
plt.scatter(cretestX_minmax[:,0], cretestX_minmax[:,1])
```

```python
np.savetxt("cretestX_minmax.csv", cretestX_minmax, delimiter=",")
np.savetxt("cretrainX_minmax.csv", cretrainX_minmax, delimiter=",")
```
