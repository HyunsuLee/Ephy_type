```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
cre = pd.read_csv('ephys_data_cre.csv')
cre.info()
```

```python
cre_dropna = cre.dropna(axis=0, how='any')
```

```python
cre_dropna.boxplot()
```

```python
def draw_kdeplot(feature): # drawing kde plot according to feature vs binary neuron class
    sns.kdeplot(cre_dropna[cre_dropna['binary_neuron'] == 'Excitatory'][feature], color='r')
    sns.kdeplot(cre_dropna[cre_dropna['binary_neuron'] == 'Inhibitory'][feature], color='b')
    return
```

```python
draw_kdeplot('height')
```

```python
draw_kdeplot('firing_rate')
```

```python
draw_kdeplot('input_resistance_mohm')
```

```python
draw_kdeplot('peak_v_ramp')
```

```python
draw_kdeplot('upstroke_downstroke_ratio_short_square')
```

```python
draw_kdeplot('latency')
```
