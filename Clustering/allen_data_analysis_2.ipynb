```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
cre = pd.read_csv('ephys_data_cre.csv')
cre.info()
```

```python
cre_dropna = cre.dropna(axis=0, how='any')
```

```python
cre_dropna.boxplot()
```

```python
sns.kdeplot(cre_dropna[cre_dropna['binary_neuron'] == 'Excitatory']['height'], color='r')
sns.kdeplot(cre_dropna[cre_dropna['binary_neuron'] == 'Inhibitory']['height'], color='b')
```

```python
sns.distplot(cre_dropna[cre_dropna['binary_neuron'] == 'Excitatory']['adaptation'], color='r')
sns.distplot(cre_dropna[cre_dropna['binary_neuron'] == 'Inhibitory']['adaptation'], color='b')
```
