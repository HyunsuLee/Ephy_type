# Preparing data, tensorflow implementation


```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cretest = pd.read_csv('cretest.csv')
cretrain = pd.read_csv('cretrain.csv')
```

```python
cretestX = cretest.drop(['Unnamed: 0', 'transgenic_line', 'reporter_status', 'binary_neuron'], axis=1)
cretrainX = cretrain.drop(['Unnamed: 0', 'transgenic_line', 'reporter_status', 'binary_neuron'], axis=1)
```

```python
cretestY = cretest['binary_neuron']
cretrainY = cretrain['binary_neuron']
```

```python
cretestX = np.array(cretestX)
cretrainX = np.array(cretrainX)
cretrainX.shape
```

```python
cretestY = np.eye(2)[cretestY]
cretrainY = np.eye(2)[cretrainY]
```

## logistic regression test

```python
def init_weights(shape):
    return tf.Variable(tf.random_normal(shape, stddev=0.01))

def model(X, w):
    return tf.matmul(X, w) 
```

```python
X = tf.placeholder("float", [None, 43]) # create symbolic variables
Y = tf.placeholder("float", [None, 2])
```

```python
w = init_weights([43, 2])
```

```python
pr_x = model(X, w)
```

```python

cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=pr_x, labels=Y)) # compute mean cross entropy (softmax is applied internally)
train_op = tf.train.GradientDescentOptimizer(0.05).minimize(cost) # construct optimizer
predict_op = tf.argmax(pr_x, 1)
```

```python
with tf.Session() as sess:
    # you need to initialize all variables
    tf.global_variables_initializer().run()

    for i in range(1000):
        for start, end in zip(range(0, len(cretrainX), 128), range(128, len(cretrainX)+1, 128)):
            sess.run(train_op, feed_dict={X: cretrainX[start:end], Y: cretrainY[start:end]})
        print(i, np.mean(np.argmax(cretestY, axis=1) ==
                         sess.run(predict_op, feed_dict={X: cretestX})))
```
